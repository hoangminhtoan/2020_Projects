```python
import numpy as np 
import os 
import matplotlib as mpl 
import matplotlib.pyplot as plt 

import torch 
import torch.nn as nn 
import torch.optim as optim

%matplotlib inline 

torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
input_t = torch.arange(-3, 3.1, 0.2)
input_t, input_t.shape
```

```python
activation_list = [
    nn.Tanh(),
    nn.Hardtanh(),
    nn.Sigmoid(),
    nn.Softmax(),
    nn.ReLU(),
    nn.LeakyReLU(negative_slope=0.1)
]

fig = plt.figure(figsize=(14, 28), dpi=600)

for i, activation in enumerate(activation_list):
    subplot = fig.add_subplot(len(activation_list), 3, i+1)
    subplot.set_title(type(activation).__name__)

    output_t = activation(input_t)

    plt.grid()
    plt.plot(input_t.numpy(), output_t.numpy(), 'k', linewidth=1)
    plt.plot([-3, 3], [0, 0], 'k', linewidth=1)
    plt.plot([0, 0], [-3, 3], 'k', linewidth=1)
    plt.plot(input_t.numpy(), output_t.numpy(), 'r', linewidth=3)

output_t

plt.show()
```
